```python
import requests
import shutil
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
url = "https://talentedme.ru/learning/?SHOWALL_1=1"
r = requests.get(url)
soup = BeautifulSoup(r.content)
boxes = soup.find_all(attrs={"class":"desc col-md-4 col-xs-12 course"})
```

```python
data = []

for box in boxes:
    header = box.find('h3').find('a').contents[0]
    link =  "https://talentedme.ru" + box.find('h3').find('a')['href']
    text = box.find('p').contents[0]
    course = {"url":link, "header":header, "text":text}
    data.append(course)
```

```python
df = pd.DataFrame.from_dict(data)
df
```

```python
df = df.drop(index=[11, 20, 35])
df
```

```python
image_links = []

images = soup.find_all(attrs={"class":"img-responsive"})
for image in images[1:-5]:
    image_links.append("https://talentedme.ru"+image['src'])
```

```python
i = 252
for image in image_links:
    r = requests.get(image, stream=True)
    if r.status_code == 200:
        with open("/content/drive/MyDrive/internship_bot/images/{}.jpg".format(i), 'wb') as f: 
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
        i += 1
```
